Iremos utilizar a biblioteca sklearn para realizar nosso modelo de regressão.

```python
from sklearn.model_selection import train_test_split

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa']

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2)
```

É importante definir quem serão o x, y e as outras variaveis, x_treino, x_teste e etc., que são os argumentos da nossa função, e os seus derivados onde através desa base relacionada, o algoritimo aprenderá.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split

x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa'] 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_rlogistica = LogisticRegression(max_iter=1000)
modelo_rlogistica.fit(x_treino, y_treino)
modelo_rlogistica.score(x_teste, y_teste)
```

Vemos aqui que ele acertou cerca de 62%, mas como saber se essa porcentagem é o suficiente para esse estudo? Vamos então comparar esse modelo com um outro mais simples.

```python
from sklearn.dummy import DummyClassifier
from sklearn.metrics import accuracy_score

modelo_dummy = DummyClassifier('most_frequent')
modelo_dummy.fit(x_treino, y_treino)
previsao_dummy = modelo_dummy.predict(x_teste)
accuracy_score(y_teste, previsao_dummy)
```

Isso quer dizer que, caso o modelo chutasse a mesma classe para todos os eventos da base de teste, teríamos um acerto de 60% dos casos, o que de certa forma condiz com o valor de ativos moa.

```python
dados_combinados['ativo_moa'].value_counts(normalize=True)
```

É pontuado em aula a importância do valor que esse número carrega, 62% em testes clinicos não é suficiente, pois poderiamos aprovar ou descatar estudos relevantes e importantes. Agora, vamos treinar uma árvore de decisão.

```python
from sklearn.tree import DecisionTreeClassifier


x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa'] 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

modelo_arvore = DecisionTreeClassifier(max_depth = 3)
modelo_arvore.fit(x_treino, y_treino)
modelo_arvore.score(x_teste, y_teste)
```

```python
from sklearn import tree

fig, ax = plt.subplots(figsize=(15, 10), facecolor='k')
tree.plot_tree(modelo_arvore,
               ax=ax,
               fontsize=10,
               rounded=True,
               filled=True,
               feature_names=x_treino.columns,
               class_names=['Não Ativado', 'Ativado'])

plt.show()
```

Para melhorar então nosso modelo que inicialmente com três camadas não obteve uma porcentagem aceitável para esse contexto, iremos variar o max_depth dentro de um intervalor de valores e comparar os resultados obtidos. 

```python
from sklearn.tree import DecisionTreeClassifier


x = dados_combinados.select_dtypes('float64')
y = dados_combinados['ativo_moa'] 
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size = 0.2, stratify=y, random_state=376)

teste=[]
treino=[]
for i in range(1,15):
    modelo_arvore = DecisionTreeClassifier(max_depth = i)
    modelo_arvore.fit(x_treino, y_treino)
    teste.append(modelo_arvore.score(x_teste, y_teste))
    treino.append(modelo_arvore.score(x_treino, y_treino))
```

O que aconteceu como nosso modelo de testes é um fenomeno chamado overfitting: "sobre-ajuste ou sobreajuste é um termo usado em estatística para descrever quando um modelo estatístico se ajusta muito bem ao conjunto de dados anteriormente observado, mas se mostra ineficaz para prever novos resultados." Vemos isso no ínicio do nosso gráfico, onde até x=2, temos ambas linhas totalmente juntas, ou seja, o treino acomapanha o modelo de testes, mas em seguida suas previsões falham destoando os resultados. 

Então usaremos o RandomForest para melhorar nossa eficácia sem gerar um overfitting. Mas antes, é necessário  realizar alguns ajustes nos dados que temos.
